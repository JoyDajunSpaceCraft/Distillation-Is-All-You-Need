```python
from langchain_community.document_loaders import JSONLoader
import json
from pathlib import Path
from pprint import pprint


file_path='./data/langchain_example.json'
bonecancer = "./knowledge/bonecancer/SubBoneCancer.json"
data = json.loads(Path(file_path).read_text())
```

```python
# loader = JSONLoader(
#     file_path='./data/langchain_example.json',
#     jq_schema='.messages[].content',
#     text_content=False)

# data = loader.load()
# from langchain.loaders import JSONLoader

# 为了加载问题和答案文本，我们需要定义一个jq_schema，它可以提取所有问题和答案
jq_schema = '.[] | .[].question.en, .[].answer.en'
bonecancer = "./knowledge/bonecancer/SubBoneCancer.json"
# 请注意，根据你的具体需求，这个jq_schema可能需要进一步调整来精确匹配你想要的数据结构
loader = JSONLoader(
    file_path=bonecancer,
    jq_schema=jq_schema,
    text_content=True  # 假设你想要以文本内容的形式加载数据
)

data = loader.load()
```

```python
pprint(data)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
# embeddings = OpenAIEmbeddings(openai_api_key="sk")
embeddings = OpenAIEmbeddings(openai_api_key="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")

documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(documents)
db = FAISS.from_documents(texts, embeddings)

retriever = db.as_retriever(
    search_type="similarity_score_threshold", search_kwargs={"score_threshold": 0.5}
)
docs = retriever.get_relevant_documents("What is the CHON-3 ?")

# retriever = db.as_retriever()

# docs = retriever.get_relevant_documents("What is the CHON-3 ?")
```
